### ※ xpath

XPath는 XML 또는 HTML 문서 내에서 특정 요소나 속성을 선택하기 위해 사용되는 경로 표현 언어입니다. 웹 크롤링이나 자동화 도구에서 주로 사용되며, 요소를 효율적으로 찾을 수 있도록 도와줍니다. 일반적인 XPath는 특정 위치나 속성을 기준으로 요소를 선택하는 상대적인 경로를 사용합니다. 또한 full xpath는 루트 요소에서 시작하여 대상 요소까지의 절대적인 경로를 나타냅니다. 따라서 문서 구조가 변경되면 경로가 깨질 가능성이 높습니다.

# **3. 픽사베이 이미지 다운로드**

```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from urllib.request import Request, urlopen
```

```python
driver = webdriver.Chrome()
url = "https://pixabay.com/images/search/snow/"
driver.maximize_window()
driver.get(url)
# time.sleep(3)
```

```python
image_xpath = "/html/body/div[1]/div[1]/div/div[2]/div[3]/div/div/div[2]/div[2]/div/a/img"
wait = WebDriverWait(driver, 10) # 최대 10초까지 기다림
image_element = wait.until(EC.presence_of_element_located((By.XPATH, image_xpath)))
```

```python
image_url = image_element.get_attribute("src")
image_byte = Request(image_url, headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/142.0.0.0 Safari/537.36'})
image_byte
```

```python
with open('snow.jpg', 'wb') as f:
    f.write(urlopen(image_byte).read())
print('파일저장 완료!!')

driver.quit()
```

# **4. 야놀자 리뷰**
- [사이트](https://nol.yanolja.com/reviews/domestic/1000111510)

```python
def crawl_yanolja_reviews(name, url, max_scroll=3):
    review_list = []
    driver = webdriver.Chrome()
    driver.get(url)
    time.sleep(3)

    last_height = driver.execute_script("return document.body.scrollHeight")
    # print(last_height)
    same_count = 0
    scroll_times = 0

    while True:
        scroll_times += 1
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
        time.sleep(2)
        new_height = driver.execute_script("return document.body.scrollHeight")

        if new_height == last_height:
            same_count += 1
        else:
            same_count = 0
            last_height = new_height
        if same_count >= 3:
            print('더 이상 새로운 리뷰가 없습니다.')
            break
        if scroll_times >= max_scroll:
            print('최대 스코롤 횟수 도달!! 종료합니다.')
            break

        # /html/body/div[1]/section/div/div[5]/div[1]
        # /html/body/div[1]/section/div/div[5]/div[2]
        
    containers = driver.find_elements(By.XPATH, "/html/body/div[1]/section/div/div[5]/div")
    # print('총 리뷰 개수: ', len(containers))

    for c in containers:
        try:
            # raw_text = c.text.strip() 
            date_text = c.find_element(By.XPATH, "./div[2]/div/div[1]/div[1]/div[2]/p").text
            comment_text = c.find_element(By.XPATH, "./div[2]/div/div[3]/div[1]/p").text
            
        except:
            date_text = ""
            comment_text = ""
            
        review_list.append({
            "review": comment_text,
            "date": date_text
        })
        
    driver.quit()
    return review_list
```

```python
reviews = crawl_yanolja_reviews("베리온", "https://nol.yanolja.com/reviews/domestic/1000111510", 100)
```
